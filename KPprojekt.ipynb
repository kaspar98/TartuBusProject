```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import codecs
import csv
```

```python
with open('varem.csv') as f:
    tekst = f.readlines()
```

```python
with open('varemkirjutatud.csv', mode='w', encoding="utf-8") as f:
    wr = csv.writer(f, delimiter=';')
    for rida in tekst:
        osad = rida.strip().split(';')
        wr.writerow(osad)
```

```python
varem = pd.read_csv('varemkirjutatud.csv', sep=";")
varem
```
